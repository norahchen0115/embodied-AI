# 一、ONNX
一个训练好的模型通常是一个个 Python 对象，只有 PyTorch 框架知道如何运行，在不支持 Python/PyTorch 的环境中，如嵌入式设备，就无法运行。为解决这个跨平台兼容性的问题，我们需要把模型以 ONNX 格式导出。

ONNX 全称 Open Neural Network Exchange（开放神经网络交换），是一套表示深度神经网络模型的开放格式。它定义了一组与环境和平台无关的标准格式，为 AI 模型的互操作性提供了基础，使 AI 模型可以在不同框架和环境下交互使用。

### Model Conversion: PyTorch to ONNX
**step 1:** set model to evaluation mode  

**step 2:** create a dummy input with the shape the model expects  

**step 3:** call `torch.onnx.export` function with a series of parameters to convert the model format

```python
model.eval()

dummy_input = torch.radn(1,3,224,224)

torch.onnx.export(
    model=model,                  # The trained PyTorch model to be exported. 
    args=dummy_input,             # A dummy input tensor with the expected shape for tracing.
    f="model.onnx",               # The name of the output .onnx file. 
    export_params=True,           # Exports the model's trained weights (parameters).
    opset_version=11,             # The ONNX version to use. Using 11 to be compatible with TF in the Optional section
    do_constant_folding=True,     # An optimization that pre-computes constant expressions in the model.
    input_names=['x', 'y', 'z'],  # Assigns names to the model's input tensors for easier identification. (might be multiple inputs)
    output_names=['prediction'],  # Assigns names to the model's output tensors. (might be multiple outputs)
    dynamic_axes=                 # Specifies which dimensions can vary in size and assign a parameter name to the dynamic axis. 
    {'input':{0:'batch_size'}, 'output':{0:'batch_size'}}
)
```
